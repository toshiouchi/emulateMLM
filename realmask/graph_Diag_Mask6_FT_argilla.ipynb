```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import json
import pickle
```

```python
with open('Diag_Mask6_FT_argilla.pkl', 'rb') as f:
    history6 = pickle.load(f)
    
len_tr_loader6 = history6["len_tr_loader"][0]
#print( len_tr_loader)
len_val_loader6 = history6["len_val_loader"][0] 
#print( len_val_loader)

#print( history6['val_loss'] )

x6 = []
y6 = []
y_acc6 = []
#print( len( history["train_loss"] ))
for i in range( len( history6["train_loss"] )):
    #print( i )
    #print( i // len_tr_loader )
    x6.append( i / len_tr_loader6 )
    y6.append( np.mean(history6['train_loss'][i-len_tr_loader6:i]) )
    y_acc6.append( np.mean(history6['train_acc'][i-len_tr_loader6:i]))

#print( y6 )    
    
v_x6 = []
v_y6 = []
v_y_acc6 = []
#print( len( history["val_loss"]) )
for i in range( len( history6["val_loss"]) ):
    v_x6.append( i / len_val_loader6 )
    v_y6.append( np.mean(history6['val_loss'][i-len_val_loader6:i]))
    v_y_acc6.append( np.mean( history6['val_acc'][i-len_val_loader6:i]))

#print( v_y6 )

plt.title( "Loss")
plt.xlabel( "Epochs")
plt.ylabel( "Loss")
plt.plot(x6, y6, label="realmask train_loss")
plt.plot( v_x6, v_y6, label="realmask val_loss")
#plt.xlim( 84000, 100000 )
#plt.ylim( 0, 2 )
plt.legend()
plt.show()
'''
plt.title( "Accuracy")
plt.xlabel( "Epochs")
plt.ylabel( "Accuracy")
plt.plot(x, history["train_acc"], label="train_acc")
plt.plot( v_x, history["val_acc"], label="val_acc")
plt.legend()
plt.show()
'''
plt.title( "WER")
plt.xlabel( "Epochs")
plt.ylabel( "WER")
plt.plot(x6, y_acc6, label="realmask train_acc")
plt.plot( v_x6, v_y_acc6 , label="realmask val_acc")
#plt.xlim( 84000, 100000 )
#plt.ylim( 0, 40 )
plt.legend()
plt.show()

#plt.title( "BLEU")
#plt.xlabel( "Epochs")
#plt.ylabel( "BLEU")
#plt.plot(x, history["train_bleu"], label="train_bleu")
#plt.plot( v_x, history["val_bleu"] , label="val_bleu")
##plt.xlim( 84000, 100000 )
#plt.ylim( 40, 120 )
#plt.legend()
#plt.show()
```
