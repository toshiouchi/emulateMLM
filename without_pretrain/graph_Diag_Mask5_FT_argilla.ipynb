```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#import json
import pickle
```

```python
with open('Diag_Mask5_FT_argilla.pkl', 'rb') as f:
    history = pickle.load(f)
    
len_tr_loader = history["len_tr_loader"][0]
#print( len_tr_loader)
len_val_loader = history["len_val_loader"][0] 
#print( len_val_loader)


x = []
y = []
y_acc = []
#print( len( history["train_loss"] ))
for i in range( len( history["train_loss"] )):
    #print( i )
    #print( i // len_tr_loader )
    x.append( i / len_tr_loader )
    y.append( np.mean(history['train_loss'][i-len_tr_loader:i]) )
    y_acc.append( np.mean(history['train_acc'][i-len_tr_loader:i]))
    
v_x = []
v_y = []
v_y_acc = []
#print( len( history["val_loss"]) )
for i in range( len( history["val_loss"]) ):
    v_x.append( i / len_val_loader )
    v_y.append( np.mean(history['val_loss'][i-len_val_loader:i]))
    v_y_acc.append( np.mean( history['val_acc'][i-len_val_loader:i]))

plt.title( "Loss")
plt.xlabel( "Epochs")
plt.ylabel( "Loss")
#plt.plot(x, history["train_loss"], label="train_loss")
#plt.plot( v_x, history["val_loss"], label="val_loss")
plt.plot( x, y, label="train_loss mean ")
plt.plot( v_x, v_y, label="val_loss mean")
             
#plt.xlim( 84000, 100000 )
plt.legend()
plt.show()
'''
plt.title( "Accuracy")
plt.xlabel( "Epochs")
plt.ylabel( "Accuracy")
plt.plot(x, history["train_acc"], label="train_acc")
plt.plot( v_x, history["val_acc"], label="val_acc")
plt.legend()
plt.show()
'''
plt.title( "ACC")
plt.xlabel( "Epochs")
plt.ylabel( "ACC")
#plt.plot(x, history["train_acc"], label="train_acc")
#plt.plot( v_x, history["val_acc"] , label="val_acc")
plt.plot( x, y_acc, label="train_acc mean ")
plt.plot( v_x, v_y_acc, label="val_acc mean")
#plt.xlim( 84000, 100000 )
plt.legend()
plt.show()

#plt.title( "BLEU")
#plt.xlabel( "Epochs")
#plt.ylabel( "BLEU")
#plt.plot(x, history["train_bleu"], label="train_bleu")
#plt.plot( v_x, history["val_bleu"] , label="val_bleu")
##plt.xlim( 84000, 100000 )
#plt.legend()
#plt.show()
```
