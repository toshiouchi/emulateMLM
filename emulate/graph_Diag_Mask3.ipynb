```python
import matplotlib.pyplot as plt
import pandas as pd
#import json
import pickle
```

```python
with open('Diag_Mask3.pkl', 'rb') as f:
    history = pickle.load(f)
    
len_tr_loader = history["len_tr_loader"][0]
print( len_tr_loader)
len_val_loader = history["len_val_loader"][0] 
print( len_val_loader)

x = []
print( len( history["train_loss"] ))
for i in range( len( history["train_loss"] )):
    #print( i )
    #print( i // len_tr_loader )
    x.append( i / len_tr_loader )
    
v_x = []
print( len( history["val_loss"]) )
for i in range( len( history["val_loss"]) ):
    v_x.append( i / len_val_loader )

#print( history["val_loss"])
    
plt.title( "Loss")
plt.xlabel( "Epochs")
plt.ylabel( "Loss")
plt.plot(x, history["train_loss"], label="train_loss")
plt.plot( v_x, history["val_loss"], label="val_loss")
#plt.xlim( 84000, 100000 )
plt.ylim( 0, 2 )
plt.legend()
plt.show()
'''
plt.title( "Accuracy")
plt.xlabel( "Epochs")
plt.ylabel( "Accuracy")
plt.plot(x, history["train_acc"], label="train_acc")
plt.plot( v_x, history["val_acc"], label="val_acc")
plt.legend()
plt.show()
'''
plt.title( "WER")
plt.xlabel( "Epochs")
plt.ylabel( "WER")
plt.plot(x, history["train_wer"], label="train_wer")
plt.plot( v_x, history["val_wer"] , label="val_wer")
#plt.xlim( 84000, 100000 )
plt.ylim( 0, 40 )
plt.legend()
plt.show()

plt.title( "BLEU")
plt.xlabel( "Epochs")
plt.ylabel( "BLEU")
plt.plot(x, history["train_bleu"], label="train_bleu")
plt.plot( v_x, history["val_bleu"] , label="val_bleu")
plt.ylim( 40, 120 )
#plt.xlim( 84000, 100000 )
plt.legend()
plt.show()
```
